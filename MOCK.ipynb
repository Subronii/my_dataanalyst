```python
toteng = mfile['ENGLISH'].value_counts()
toteng.plot(kind='barh', color=sns.color_palette("Spectral", len(toteng)))
plt.title('RANKS IN ENGLISH')
plt.xlabel('SCORE IN ENGLISH')
plt.show()
```

```python
plt.figure(figsize=(12, 10))
toteng = mfile['ENGLISH'].value_counts()
topeng = toteng.head(20)
topeng.plot(kind='barh', color=sns.color_palette("Spectral", len(toteng)))
plt.title('TOP SCORER IN ENGLISH')
plt.xlabel('SCORER IN ENGLISH', fontsize=14)
plt.ylabel('SCORE', fontsize=14)
plt.xticks(fontsize=12)
plt.yticks(fontsize=12)
plt.show()
```
