```python
import matplotlib.pyplot as plt
import numpy as np

# Sample data
x = np.linspace(0, 10, 100)
y1 = np.sin(x)
y2 = np.cos(x)
y3 = np.tan(x)

# Create a figure
plt.figure(figsize=(12, 8))

# First subplot (1 row, 3 columns, 1st plot)
plt.subplot(1, 3, 1)
plt.plot(x, y1, color='blue')
plt.title('Sine Function')
plt.xlabel('x')
plt.ylabel('sin(x)')

# Second subplot (1 row, 3 columns, 2nd plot)
plt.subplot(1, 3, 2)
plt.plot(x, y2, color='orange')
plt.title('Cosine Function')
plt.xlabel('x')
plt.ylabel('cos(x)')

# Third subplot (1 row, 3 columns, 3rd plot)
plt.subplot(1, 3, 3)
plt.plot(x, y3, color='green')
plt.title('Tangent Function')
plt.xlabel('x')
plt.ylabel('tan(x)')
plt.ylim(-10, 10)  # Limit y-axis for better visualization

# Adjust layout
plt.tight_layout()

# Show the plots
plt.show()
```
